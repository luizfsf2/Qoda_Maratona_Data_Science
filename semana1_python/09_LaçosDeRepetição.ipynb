# **>>> WHILE**

---



```python
# while
# ENQUANTO A DECLARAÇÃO FOR True, O BLOCO SERÁ EXECUTADO 

n = 0
while n <= 10:
  print(n)
  n = n+1  # n += 1
```

```python
# +=, -=, *=, /= ...
n = 0
while n <= 10:
  print(n)
  n += 1 
```

```python
# while True realiza o código por tempo indeterminado: cuidado!
while True:
    print(0)
```

## **O break**

---



```python
# Com a instrução break podemos parar o loop mesmo se a condição while for verdadeira
n = 0
while n <= 10:
    print(n)
    if n == 3:
        break
    n = n+1 
```

## **O continue**

---



```python
# Com a instrução continue podemos continuar a próxima iteração de forma ilimitada: cuidado!
n = 0
while n <= 10:
    print(n)
    if n == 3:
        continue
    n = n+1 
```

# **>>> FOR**

---



```python
# ITERAÇÃO ATRAVÉS DE SEQUÊNCIAS (lista, tupla, dicionario, set, string)

pessoas = ['Ana', 'Maria', 'Alita']
for pessoa in pessoas:
    print(pessoa)
```

```python
# ITERAÇÃO EM STRINGS

x = "Alita"
for letra in x:
    print(letra)
```

```python
# ITERAÇÃO RANGE
for x in range(10):
    print(x)
```

```python
# TAMBÉM PODE RECEBER "INICIO,FIM,PASSO"
for x in range(0,100,2):
    print(x)
```

```python
# PODERÁ RECEBER ELSE
for x in range(10):
    print(x)
else:
    print("fim!")
```

```python
# PODERÁ RECEBER break 
for x in range(10):
    print(x)
    if x == 5:
        print("fim!")
        break
```

```python
# E continue
for x in range(10):    
    if x == 5:
        print("Pula!")
        continue
    print(x)    
```

```python
# Loops Aninhados
a = ['verde', 'roxa', 'amarela']
b = ['maçã', 'uva', 'banana']

for x in a:
    for y in b:
        print(x,y)
```

1.   Crie um algoritmo que solicite ao usuário 3 números e printe na tela a média destes (fácil)



```python
numeros = []
for i in range(1,4):
    n = int(input(f'Número {i}: '))
    numeros.append(n)

soma = 0
for n in numeros:
    soma += n

print(f'Números: {numeros}')
print(f'Média: {soma / len(numeros):.2f}')
```

2.   Crie um algoritmo que solicite ao usuário um número e apresente o fatorial deste número como resultado (médio)
[Clique aqui para saber como calcular o fatorial.](https://pt.wikipedia.org/wiki/Fatorial)


```python
n = int(input('Número: '))

fatorial = 1
i = n
while i > 0:
    fatorial *= i
    i -= 1

print(f'{n}!= {fatorial}')    
```

3.   Crie um algoritmo que solicite 5 números. Ao término, printar o maior e o menor número (difícil)

```python
ns = []
for i in range(1,6):
    n = int(input(f'Número {i}: '))
    ns.append(n)
ns.sort()
print(f'O maior é {ns[-1]} e o menor é {ns[0]}')
```
